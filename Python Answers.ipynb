```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1	Write a code to import  Dataframe in Python

df = pd.read_excel('F:\Datasets\Python Data.xlsx')
```

```python
df
```

```python
# 2	Write the code to find unique Name in above table

unique_names = df['Name'].unique()
```

```python
unique_names
```

```python
# 3	Write the code to find top 3 sales with Name and Department

top_3_sales = df.nlargest(3, 'Sales')[['Sales', 'Name', 'Department']]
```

```python
top_3_sales
```

```python
# 4	Replace "DeRusha, Joe" Name with "John" in D Column

df['Name'] = df['Name'].replace('DeRusha, Joe', 'John')
```

```python
df['Name']
```

```python
df
```

```python
# 5. 	Plot top 5 sales for each Region

df = df.sort_values(by=['Region', 'Sales'], ascending=[True, False])

top_5_sales = df.groupby('Region').head(5)

plt.figure(figsize=(10, 6))
for region, data in top_5_sales.groupby('Region'):
    plt.bar(data['Name'], data['Sales'], label=region)

plt.xlabel('Name')
plt.ylabel('Sales')
plt.title('Top 5 Sales for Each Region')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 6	Plot the chart -> percentage contribution of sales for Unique Name

name_sales = df.groupby('Name')['Sales'].sum().reset_index()
```

```python
name_sales
```

```python
# Sales% through Python

total_sales = name_sales['Sales'].sum()
name_sales['Percentage'] = (name_sales['Sales'] / total_sales) * 100
```

```python
name_sales['Percentage']
```

```python
name_sales = name_sales.sort_values(by='Percentage', ascending=False)
```

```python
name_sales
```

```python
plt.figure(figsize=(10, 6))
plt.bar(name_sales['Name'], name_sales['Percentage'])
plt.xlabel('Name')
plt.ylabel('Percentage Contribution (%)')
plt.title('Percentage Contribution of Sales for Unique Names')
plt.xticks(rotation=45)
plt.show()
```
